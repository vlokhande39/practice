```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/tips.csv")
```

```python
data.head()
```

# we are using tips data to studying about seaborn library for data analysis

```python
#scatter plot in between total_bill and tips
#one type of scatter plot is relplot
#
sns.relplot(x='total_bill', y = 'tip', data=data)
#we can see in the graph their is a linear relationships in the two attributes.
```

```python
#adding hue parameter -as s smoker
sns.relplot(x='total_bill', y='tip', hue='smoker', data=data)
#we are checking the distribution as a hue parameter smoker
```

```python
#style paramter
sns.relplot(x='total_bill',y='tip', hue='smoker', style='sex',data=data)
#by that kind of plots we can easily see the how many smokers are male and female, and what kind for relationship of that three attributes at a time.
```

```python
#size parameter
sns.relplot(x='total_bill',y='tip', hue='smoker', style='sex', size='size', sizes=(15,200), data=data)
#we can change the sizes of the parameter.
```

```python
#catergorial scatterplot
#x = catergorical data, y= continuous data.
sns.catplot(x='day', y='tip',kind='strip',data=data, jitter=0)
#it is strip plots.jitter=it is used the noise in between two points.
```

```python
sns.catplot(x='day', y='tip', kind='swarm', data=data)
#in the swarm type of plot we can easily see the count is increased on the basis of repetative values.
#we can see the distribution of the data
```

```python
sns.catplot(x='day', y ='tip', kind='swarm', hue='sex', data=data)
```

```python
sns.swarmplot(x='day', y='tip',data=data) #instead of catplot we can directly use the swarmplot.
```

```python
#boxplot- a boxplot is a standardized the value based on the 5 number summary. Q1,Q2,Q3,IQR,median.
#basically it is used to detect the outliers in the datasets.
#univariate analysis
sns.boxplot(data['tip'])
#we can check the outliers presents in the data, 50% data, median values and all
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=data)
#boxplots with the bases of daywise.
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='box',data=data)
#we can check with the gender also ourboxplot
```

```python
#violin plot plays a similar role as a box and whisker plot.
#It shows the distribution of quantitative data across several levels of one (or more) categorical.
#advantage ot violin plots is, not only it's giving the 5 number summery but also it's showing the entire distribution of the data.
sns.violinplot(data['total_bill']) #both side are same.
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=data)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split='True',data=data)#distribution of male and female with day attributtes.
```

```python
#barplot.
#central tendency
sns.catplot(x='smoker',y='total_bill',kind='bar',data=data)#mean of smoker is higher than non smoker.
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',estimator=np.var,data=data)
```
